```python
from random import  randint
import timeit
from timeit import repeat

ARRAY_LENGTH = 10000

array = [randint(0, 1000) for i in range(ARRAY_LENGTH)]
print(array)
times = timeit.timeit(stmt= f"sorted({array})",number = 1)
print(f"เวลาที่ใช้:{times}")
```

```python
from random import  randint
from timeit import repeat

ARRAY_LENGTH = 10000

array = [randint(0, 1000) for i in range(ARRAY_LENGTH)]
print(array)
times = repeat(stmt= f"sorted({array})",number = 1)
print(f"เวลาที่ใช้ที่ใช้น้อยที่สุด:{min(times)}")
```
